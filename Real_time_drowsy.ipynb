```python
import cv2
```

```python
import numpy as np
import tensorflow as tf
```

```python
model_path = r"C:\Users\ASUS\Desktop\Drowsy\Model"
model = tf.keras.models.load_model(model_path)
ddd_class=['Drowsy', 'Non Drowsy']
warning=[0]
```

```python
def predict(model, img):
    img=tf.expand_dims(img,0)
    predictions=model.predict(img)
    print("0: ",predictions[0][0])
    print("1: ",predictions[0][1])
    if predictions[0][0]>0.0001:
        index=0
        predicted_class=ddd_class[0]
        confidence =round(100*(predictions[0][0]),2)
        return index,predicted_class, confidence
    else:
        index=np.argmax(predictions[0])
        predicted_class=ddd_class[np.argmax(predictions[0])]
        confidence =round(100*(np.max(predictions[0])),2)
        return index,predicted_class, confidence
```

```python
def predict(model, img):
    img=tf.expand_dims(img,0)
    predictions=model.predict(img)
    print(predictions)
    index=np.argmax(predictions[0])
    predicted_class=ddd_class[np.argmax(predictions[0])]
    confidence =round(100*(np.max(predictions[0])),2)
    return index,predicted_class, confidence
```

```python
cap = cv2.VideoCapture(r"C:\Users\ASUS\Downloads\Twitch streamer starts to fall asleep while driving....mp4")

while True:
    ret, live = cap.read() 
    
    if not ret:
        break
    
    frame=tf.keras.preprocessing.image.smart_resize(live,(227,227))
    live=frame/255
    index,predicted_class,confidence = predict(model,frame)
    
    if index in warning:
        cv2.putText(live,f"ALERT!!!{predicted_class}", (10,30),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    else:
        cv2.putText(live,f"{predicted_class}", (10,30),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        
    print({predicted_class})   
    cv2.imshow('Disease Detection', live)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
